```python
# 2.Create a Bus child class that inherits from the Vehicle class. The default fare charge of any vehicle is seating capacity * 100. If Vehicle is Bus instance, we need to add an extra 10% on full fare as a maintenance charge. So total fare for bus instance
# will become the final amount = total fare + 10% of the total fare.Note: The bus seating capacity is 50. so the final fare amount should be 5500.  You need to override the fare() method of a Vehicle class in Bus class. (5)


print("Name - Shinde Smita Shahaji  \n UID = 20BCS4643")

class Vehicle:
  #construstor to initialize the object's attributes.
    def __init__(self, name, mileage, capacity):   
        self.name = name #creating instance "name" for the given class
        self.mileage = mileage #creating instance "mileage" for the given class
        self.capacity = capacity #creating instance "capacity" for the given class

    def fare(self):
        return self.capacity * 100 #returning  the capacity value

class Bus(Vehicle):
  #defining function "fare" to calculate  the value of fare_car and total_fare
    def fare(self): 
        amount = super().fare()#calculating the fare_care value
        amount += amount * 10 / 100  # calculating the total_fave value
        return amount #returning  the total fare value


Red_Bus = Bus("Red Bus", 12, 50)  #passing the values of name , mileage and capacity to main function
print("Total Bus fare is:", Red_Bus.fare()) #printing the function output





```
