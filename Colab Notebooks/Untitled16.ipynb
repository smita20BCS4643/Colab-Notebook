```python
import re

opn = open('ait.txt')
for line in opn:
    line = line.rstrip()
if re.findall('^X\S*:([0-9.]+)',line):
    print(line)
```

```python
sum = 0
count = 0
average = 0
while(True):
  try:
    line = input("enter a number")
    if(line=="done"):
      break
      value = float(line)
      sum = value+sum
      count = count+1
      average = sum/count
  except:
    print("invalid input")
print(sum,count,average)
```
