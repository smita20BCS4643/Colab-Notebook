```python
dataset = pd.read_csv('Data.csv')
#features and depending variables
# depending variables are last column 
#usually first column features 
x  = dataset.iloc[: , :-1].values
# iloc - loctate for indexes for rows and columns  finding loction on the basis of index
# : means range in python
# : all rows  and :-1 for columns
# x for features 
y = dataset.iloc[: , -1].values
```

Taking care of mising data


```python
#ignore them or remove them
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
```
