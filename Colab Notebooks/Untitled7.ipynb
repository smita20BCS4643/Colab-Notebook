```python
f=open('Book1.csv','r')

lst=[]

for i in f.read().splitlines():

    try:

            if i.lstrip('-').isdigit():

                i=int(i)

                lst.append(i)

            elif i.isdigit():

                    i=int(i)

                    lst.append(i)

    except:

        continue

    print(lst)

def findTriplets(lst, n):

    found = False

    for i in range(0, n-2):

        for j in range(i+1, n-1):

            for k in range(j+1, n):

                if (lst[i] + lst[j] + lst[k] == 0):

                    print(lst[i], lst[j], lst[k])

                    found = True

    if (found == False):

        print(" not exist ")

n = len(lst)

findTriplets(lst, n)
```
